```python
import json, requests, threading, queue, time, string
from datetime import datetime, timedelta
import asyncio

thread_limit = 600
user_snps = []

with open("ValeskaCombinedDNA.txt", "r") as fd:
    for line in fd:
        user_snps.append(line)

async def query_opensnp(rsids, index, request_count):
    with open("good_rsids-{}.txt".format(index), 'w') as fd:
        try:
            opensnp_url = "https://opensnp.org/snps/json/annotation/" + ",".join(rsids) + ".json"
            request_count += 1
            # Fetch information using requests module
            # and parse with json module
            response = requests.Session().get(opensnp_url)
            data = json.loads(response.text)
            if "error" in data:
                error_rsid = data['error'].split(" ")[-1].replace("\n", "")
                try:
                    rsids.remove(error_rsid)
                except Exception as e:
                    print(e)
                    time.sleep(10)
                return query_opensnp(rsids, fd, request_count)
            good_results = 0
            for rsid in data:
                if "snp" in data:
                    print("THIS IS BAD! ONLY ONE RESULT RETURNED! FIGURE THIS OUT! RSID_LIST:", rsids)
                    if len(data["snp"]["annotations"]["snpedia"]) > 0:
                        fd.write(rsid + "\n")
                        good_results += 1
                else:
                    if len(data[rsid]["annotations"]["snpedia"]) > 0:
                        fd.write(rsid + "\n")
                        good_results += 1
            print("Batch of ", len(rsids), " completed!", "Matching results in opensnp:", good_results, " Percentage match in opensnp:", (good_results/len(rsids))*100)
            return good_results, request_count
        except Exception as e:
            print("ERROR: Last rsid was:", rsids[-1])
            print(e)


async def fetch_bulk_RSID_info(rsids):
    
    # Enqueue list of RSIDS that need to be queried
    rsid_list = []
    row = []
    for i in range(0, len(rsids)):
        if i % thread_limit == 0 and i != 0:
            rsid_list.append(row)
            row = []
        rsid = rsids[i]
        row.append(rsid.split("\t")[0].replace("\n", ""))
    print(len(rsid_list), " RSID lookups enqued")
    #print(rsid_list[0])
    total_good_results = 0
    total_rsid_count = 0
    start_time = datetime.now()
    request_count = 0
    
    i = 0
    workers = []
    for rsids in rsid_list:
        i += 1
        workers.append(asyncio.create_task(query_opensnp(rsids, i, 0)))
        total_rsid_count += len(rsids)
        
    values = await asyncio.gather(*workers)
    for value in values:
        total_good_results += value[0]
        request_count += value[1]
        
    print("Total count:", total_good_results, " total percentage:", (total_good_results/total_rsid_count)*100)
    end_time = datetime.now()
    print("Total time elapsed:", end_time-start_time, " time per request:", (end_time-start_time)/request_count)

await fetch_bulk_RSID_info(user_snps)
```

```python
def parse_user_data(filename):
    rsids = dict()
    
    with open(filename) as f:
        i = 0
        while True:
            line = f.readline()
            # Skip header line
            if not line.startswith('#'):
                break
        
        # Begin to parse the real data
        while line:
            data = line.split()
            rsids[data[0]] = data[1::]
            line = f.readline()

        return rsids
```
